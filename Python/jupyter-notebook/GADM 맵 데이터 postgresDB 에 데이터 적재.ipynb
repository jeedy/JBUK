````python
import wget

for n in nation:
    print("%s ," %n)
    url = "https://biogeo.ucdavis.edu/data/gadm3.6/gpkg/gadm36_%s_gpkg.zip" %n
    wget.download(url, "./nations")
````

````python
path_to_zip = './nations'
directory_to_extract_to = './nations_unzip'

import os
import zipfile

files = os.listdir(path_to_zip)
for f in files:
    if f.endswith(".zip"):
        zip_file = path_to_zip+'/'+f
        print(zip_file)
        with zipfile.ZipFile(zip_file, 'r') as zip_ref:
            zip_ref.extractall(directory_to_extract_to)
````
